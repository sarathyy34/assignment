```python
from bs4 import BeautifulSoup 
```

```python
import requests
```

```python
url="https://www.91mobiles.com/top-10-mobiles-in-india"
```

```python
response=requests.get(url)
```

```python
response
```

```python
soup=BeautifulSoup(response.text)
```

```python
soup
```

```python
name=soup.select(".name")
```

```python
name
```

```python
processor=soup.select(".specs_li:nth-child(1) label:nth-child(3)")
```

```python
processor
```

```python
Battery=soup.select(".specs_li:nth-child(4) label:nth-child(1)")
```

```python
Battery
```

```python
price=soup.select(".price_float , .price_padding")
```

```python
price
```

```python
import pandas
```

```python
mobilelist=pandas.DataFrame(name,columns=['Mobile name'])
```

```python
mobilelist
```

```python
mobilelist["processor"]=processor
```

```python
mobilelist["Battery"]=Battery
```

```python
mobilelist["price"]=price
```
